```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the uploaded file
df = pd.read_csv('netflix_titles.csv')

# Display first 5 rows
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df.dropna(subset=['date_added', 'country'], inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title('Content Added to Netflix by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
df['listed_in'].value_counts().head(5).plot(kind='bar', color='orange')
plt.title('Top 5 Netflix Genres')
plt.xlabel('Genre Category')
plt.ylabel('Count')
plt.show()
```
